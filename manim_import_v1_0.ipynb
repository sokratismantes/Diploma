```python
%%writefile monotonic_stack.py
from manim import *


# manim -pqh monotonic_stack.py MonotonicStackQ1Tutorial
# Για πιο γρήγορο preview:
# manim -pql monotonic_stack.py MonotonicStackQ1Tutorial


class MonotonicStackQ1Tutorial(Scene):
    def construct(self):
        self.camera.background_color = "#0f1117"

        title_color = BLUE_B
        accent = YELLOW
        good = GREEN_B
        warn = ORANGE
        bad = RED_B
        text_color = GREY_A

        # -----------------------------
        # Layout helpers
        # -----------------------------
        LEFT_ZONE_X = -3.45
        RIGHT_ZONE_X = 4.15
        CONTENT_Y = -0.2
        STATUS_Y = -3.35
        ALERT_Y = -2.55

        def shrink_to_width(mobj, max_width):
            if mobj.width > max_width:
                mobj.scale(max_width / mobj.width)
            return mobj

        def shrink_to_height(mobj, max_height):
            if mobj.height > max_height:
                mobj.scale(max_height / mobj.height)
            return mobj

        def place_left(mobj, max_width=5.35, y=CONTENT_Y):
            shrink_to_width(mobj, max_width)
            mobj.move_to([LEFT_ZONE_X, y, 0])
            return mobj

        def place_bottom(mobj, max_width=11.0, y=STATUS_Y):
            shrink_to_width(mobj, max_width)
            mobj.move_to([0, y, 0])
            return mobj

        def make_status(text, color=text_color, font_size=22):
            status = Text(
                text,
                font="DejaVu Sans",
                font_size=font_size,
                color=color,
            )
            shrink_to_width(status, 11.0)
            status.move_to([0, STATUS_Y, 0])
            return status

        def make_alert(text, color=warn, font_size=24):
            alert = Text(
                text,
                font="DejaVu Sans",
                font_size=font_size,
                color=color,
            )
            shrink_to_width(alert, 10.5)
            alert.move_to([0, ALERT_Y, 0])
            return alert

        # -----------------------------
        # Visual helpers
        # -----------------------------
        def section_title(main_text, sub_text="", color=title_color):
            main = Text(
                main_text,
                font="DejaVu Sans",
                weight=BOLD,
                font_size=42,
                color=color,
            )
            sub = Text(
                sub_text,
                font="DejaVu Sans",
                font_size=22,
                color=text_color,
            )
            group = VGroup(main, sub).arrange(DOWN, aligned_edge=LEFT, buff=0.16)
            group.to_edge(UP)
            shrink_to_width(group, 11.2)
            return group

        def bullet_list(items, font_size=24, color=text_color, buff=0.26):
            bullets = VGroup()
            for item in items:
                line = Text(
                    f"• {item}",
                    font="DejaVu Sans",
                    font_size=font_size,
                    color=color,
                )
                shrink_to_width(line, 5.3)
                bullets.add(line)
            bullets.arrange(DOWN, aligned_edge=LEFT, buff=buff)
            return bullets

        def make_array_row(values, y=1.0):
            cells = VGroup()
            labels = VGroup()

            for i, v in enumerate(values):
                box = Square(side_length=0.78, color=GREY_B).set_stroke(width=2)
                box.move_to(LEFT * 2.8 + RIGHT * i * 0.88 + UP * y)

                txt = Text(
                    str(v),
                    font="DejaVu Sans",
                    font_size=26,
                    color=WHITE,
                ).move_to(box.get_center())

                idx = Text(
                    str(i),
                    font="DejaVu Sans",
                    font_size=17,
                    color=GREY_A,
                ).next_to(box, DOWN, buff=0.10)

                cells.add(box)
                labels.add(VGroup(txt, idx))

            group = VGroup(cells, labels)
            shrink_to_width(group, 5.8)
            group.move_to([-2.45, 0.9, 0])

            return cells, labels

        def make_stack_group(stack_values, origin=RIGHT * 4.25 + DOWN * 1.1):
            stack = VGroup()

            for i, val in enumerate(stack_values):
                rect = RoundedRectangle(
                    corner_radius=0.08,
                    width=1.38,
                    height=0.56,
                    color=good,
                ).set_fill(good, opacity=0.12)

                rect.move_to(origin + UP * i * 0.66)

                txt = Text(
                    str(val),
                    font="DejaVu Sans",
                    font_size=22,
                    color=WHITE,
                ).move_to(rect.get_center())

                stack.add(VGroup(rect, txt))

            return stack

        def stack_frame(height=4.15):
            frame = RoundedRectangle(
                corner_radius=0.12,
                width=1.85,
                height=height,
                color=GREY_B,
            ).set_stroke(width=2)

            frame.move_to(RIGHT * 4.25 + DOWN * 0.2)

            label = Text(
                "Stack",
                font="DejaVu Sans",
                font_size=23,
                color=accent,
            ).next_to(frame, UP, buff=0.18)

            top_label = Text(
                "top",
                font="DejaVu Sans",
                font_size=18,
                color=accent,
            ).next_to(frame.get_top(), RIGHT, buff=0.12)

            return VGroup(frame, label, top_label)

        # -----------------------------
        # Intro
        # -----------------------------
        intro = VGroup(
            Text(
                "Φθίνουσα Μονοτονική Στοίβα",
                font="DejaVu Sans",
                font_size=46,
                weight=BOLD,
                color=title_color,
            ),
            Text(
                "concept • pattern • trace • template • warnings",
                font="DejaVu Sans",
                font_size=24,
                color=text_color,
            ),
        ).arrange(DOWN, buff=0.25)

        shrink_to_width(intro, 11.2)

        self.play(FadeIn(intro, shift=UP), run_time=1.4)
        self.wait(2.0)
        self.play(FadeOut(intro), run_time=0.8)

        # ==========================================================
        # 1) CONCEPT
        # ==========================================================
        t1 = section_title(
            "concept",
            "Τι είναι η φθίνουσα μονοτονική στοίβα",
        )
        self.play(FadeIn(t1, shift=UP), run_time=0.8)

        concept_left = bullet_list(
            [
                "Στοίβα όπου από κάτω προς τα πάνω οι τιμές μένουν σε φθίνουσα σειρά.",
                "Πριν βάλουμε νέο x, πετάμε από την κορυφή όσα χαλάνε τη φθίνουσα τάξη.",
                "Χρήσιμη για κοντινότερο μεγαλύτερο ή για να κρατάμε μόνο υποψήφιους.",
            ],
            font_size=23,
        )
        place_left(concept_left, max_width=5.25, y=-0.15)

        values = [9, 7, 5]
        sf = stack_frame()
        sg = make_stack_group(values)

        invariant_box = RoundedRectangle(
            corner_radius=0.12,
            width=5.0,
            height=1.05,
            color=accent,
        ).set_fill(accent, opacity=0.08)

        invariant_box.move_to([-3.35, -2.35, 0])

        invariant_text = VGroup(
            Text(
                "Invariant:",
                font="DejaVu Sans",
                font_size=24,
                color=accent,
                weight=BOLD,
            ),
            Text(
                "κάθε νέα κορυφή ≤ την προηγούμενη",
                font="DejaVu Sans",
                font_size=21,
                color=WHITE,
            ),
        ).arrange(DOWN, aligned_edge=LEFT, buff=0.08)

        shrink_to_width(invariant_text, 4.65)
        invariant_text.move_to(invariant_box.get_center())

        self.play(
            FadeIn(concept_left, shift=RIGHT),
            Create(sf),
            FadeIn(sg, shift=UP),
            run_time=1.6,
        )
        self.play(FadeIn(invariant_box), Write(invariant_text), run_time=1.0)
        self.wait(5.0)

        x_label = Text(
            "Νέο στοιχείο: 6",
            font="DejaVu Sans",
            font_size=25,
            color=accent,
        )
        x_label.move_to([1.35, 1.3, 0])

        x_box = Square(side_length=0.78, color=accent).set_fill(accent, opacity=0.1)
        x_num = Text("6", font="DejaVu Sans", font_size=26, color=WHITE).move_to(x_box)
        x_group = VGroup(x_box, x_num).next_to(x_label, DOWN, buff=0.22)

        pop_note = make_alert("6 > 5  →  pop")

        self.play(FadeIn(x_label), FadeIn(x_group), run_time=0.8)
        self.play(
            FadeIn(pop_note),
            Indicate(sg[-1][0], color=warn),
            run_time=1.0,
        )
        self.play(FadeOut(sg[-1], shift=UP), run_time=0.7)

        old_stack_without_top = make_stack_group([9, 7])
        new_stack = make_stack_group([9, 7, 6])

        self.play(ReplacementTransform(sg[:-1], old_stack_without_top), run_time=0.5)
        self.play(FadeIn(new_stack[-1], shift=UP), run_time=0.6)
        self.play(FadeOut(pop_note), run_time=0.3)
        self.wait(3.5)

        self.play(
            FadeOut(concept_left),
            FadeOut(sf),
            FadeOut(old_stack_without_top),
            FadeOut(new_stack[-1]),
            FadeOut(invariant_box),
            FadeOut(invariant_text),
            FadeOut(x_label),
            FadeOut(x_group),
            FadeOut(t1),
            run_time=1.0,
        )

        # ==========================================================
        # 2) PATTERN
        # ==========================================================
        t2 = section_title(
            "pattern",
            "Ποια είναι η γενική ροή",
        )
        self.play(FadeIn(t2, shift=UP), run_time=0.8)

        flow_box = RoundedRectangle(
            corner_radius=0.15,
            width=10.5,
            height=4.5,
            color=BLUE_D,
        ).set_fill(BLUE_D, opacity=0.08)

        flow_box.shift(DOWN * 0.25)

        steps = VGroup(
            Text(
                "1. Σαρώνω τον πίνακα από αριστερά προς τα δεξιά",
                font="DejaVu Sans",
                font_size=26,
                color=WHITE,
            ),
            Text(
                "2. Όσο η κορυφή παραβιάζει τον κανόνα, κάνω pop",
                font="DejaVu Sans",
                font_size=26,
                color=WHITE,
            ),
            Text(
                "3. Με ό,τι μένει στην κορυφή, βγάζω πληροφορία αν χρειάζεται",
                font="DejaVu Sans",
                font_size=25,
                color=WHITE,
            ),
            Text(
                "4. Βάζω το τρέχον στοιχείο στη στοίβα",
                font="DejaVu Sans",
                font_size=26,
                color=WHITE,
            ),
        ).arrange(DOWN, aligned_edge=LEFT, buff=0.38)

        shrink_to_width(steps, 9.6)
        steps.move_to(flow_box.get_center())

        mini_hint = Text(
            "Για φθίνουσα στοίβα: αφαιρώ μικρότερα ή ίσα, ανάλογα με το πρόβλημα.",
            font="DejaVu Sans",
            font_size=23,
            color=accent,
        )
        place_bottom(mini_hint, y=STATUS_Y)

        self.play(Create(flow_box), run_time=0.8)

        for s in steps:
            self.play(Write(s), run_time=0.7)
            self.wait(1.4)

        self.play(FadeIn(mini_hint), run_time=0.7)
        self.wait(5.0)

        self.play(
            FadeOut(flow_box),
            FadeOut(steps),
            FadeOut(mini_hint),
            FadeOut(t2),
            run_time=0.9,
        )

        # ==========================================================
        # 3) TRACE
        # ==========================================================
        t3 = section_title(
            "trace",
            "Πώς εξελίσσεται σε ένα παράδειγμα",
        )
        self.play(FadeIn(t3, shift=UP), run_time=0.8)

        arr = [7, 5, 6, 3, 4, 2]
        cells, labels = make_array_row(arr, y=1.0)
        sf2 = stack_frame()

        arr_title = Text(
            "Παράδειγμα εισόδου",
            font="DejaVu Sans",
            font_size=23,
            color=accent,
        ).next_to(cells, UP, buff=0.22)

        self.play(
            FadeIn(arr_title),
            LaggedStart(*[Create(c) for c in cells], lag_ratio=0.08),
            LaggedStart(*[FadeIn(l) for l in labels], lag_ratio=0.08),
            Create(sf2),
            run_time=1.4,
        )

        current_marker = Triangle(color=accent, fill_opacity=1).scale(0.16).rotate(PI)
        current_marker.next_to(cells[0], UP, buff=0.10)

        status = make_status("Ξεκινάμε με κενή στοίβα")

        stack_vals = []
        stack_mobj = VGroup()

        self.play(FadeIn(current_marker), FadeIn(status), run_time=0.8)

        def redraw_stack():
            return make_stack_group(stack_vals)

        # -----------------------------
        # Step 0: 7
        # -----------------------------
        self.play(Indicate(cells[0], color=accent), run_time=0.6)

        stack_vals.append(7)
        new_stack = redraw_stack()

        self.play(FadeIn(new_stack, shift=UP), run_time=0.7)
        stack_mobj = new_stack

        new_status = make_status("7 μπαίνει στη στοίβα")
        self.play(Transform(status, new_status), run_time=0.6)
        self.wait(1.2)

        # -----------------------------
        # Step 1: 5
        # -----------------------------
        self.play(
            current_marker.animate.next_to(cells[1], UP, buff=0.10),
            run_time=0.5,
        )
        self.play(Indicate(cells[1], color=accent), run_time=0.5)

        stack_vals.append(5)
        new_stack = redraw_stack()

        self.play(ReplacementTransform(stack_mobj, new_stack), run_time=0.7)
        stack_mobj = new_stack

        new_status = make_status(
            "5 ≤ 7, άρα κρατάμε τη φθίνουσα σειρά και κάνουμε push"
        )
        self.play(Transform(status, new_status), run_time=0.7)
        self.wait(1.6)

        # -----------------------------
        # Step 2: 6
        # -----------------------------
        self.play(
            current_marker.animate.next_to(cells[2], UP, buff=0.10),
            run_time=0.5,
        )
        self.play(Indicate(cells[2], color=accent), run_time=0.5)

        pop_text = make_alert("6 > 5, οπότε το 5 φεύγει")

        self.play(
            FadeIn(pop_text),
            Indicate(stack_mobj[-1][0], color=warn),
            run_time=0.8,
        )

        stack_vals.pop()
        new_stack = redraw_stack()

        self.play(ReplacementTransform(stack_mobj, new_stack), run_time=0.7)
        stack_mobj = new_stack

        stack_vals.append(6)
        newer_stack = redraw_stack()

        self.play(ReplacementTransform(stack_mobj, newer_stack), run_time=0.7)
        stack_mobj = newer_stack

        self.play(FadeOut(pop_text), run_time=0.3)

        new_status = make_status(
            "Μετά το pop, το 6 μπαίνει και η στοίβα ξαναγίνεται φθίνουσα"
        )
        self.play(Transform(status, new_status), run_time=0.7)
        self.wait(1.8)

        # -----------------------------
        # Step 3: 3
        # -----------------------------
        self.play(
            current_marker.animate.next_to(cells[3], UP, buff=0.10),
            run_time=0.5,
        )
        self.play(Indicate(cells[3], color=accent), run_time=0.5)

        stack_vals.append(3)
        new_stack = redraw_stack()

        self.play(ReplacementTransform(stack_mobj, new_stack), run_time=0.7)
        stack_mobj = new_stack

        new_status = make_status("3 ≤ 6, άρα απλώς μπαίνει πάνω από το 6")
        self.play(Transform(status, new_status), run_time=0.6)
        self.wait(1.4)

        # -----------------------------
        # Step 4: 4
        # -----------------------------
        self.play(
            current_marker.animate.next_to(cells[4], UP, buff=0.10),
            run_time=0.5,
        )
        self.play(Indicate(cells[4], color=accent), run_time=0.5)

        pop_text2 = make_alert("4 > 3, άρα το 3 βγαίνει")

        self.play(
            FadeIn(pop_text2),
            Indicate(stack_mobj[-1][0], color=warn),
            run_time=0.8,
        )

        stack_vals.pop()
        new_stack = redraw_stack()

        self.play(ReplacementTransform(stack_mobj, new_stack), run_time=0.7)
        stack_mobj = new_stack

        stack_vals.append(4)
        newer_stack = redraw_stack()

        self.play(ReplacementTransform(stack_mobj, newer_stack), run_time=0.7)
        stack_mobj = newer_stack

        self.play(FadeOut(pop_text2), run_time=0.3)

        new_status = make_status("Μετά το pop, το 4 μπαίνει πάνω από το 6")
        self.play(Transform(status, new_status), run_time=0.7)
        self.wait(1.7)

        # -----------------------------
        # Step 5: 2
        # -----------------------------
        self.play(
            current_marker.animate.next_to(cells[5], UP, buff=0.10),
            run_time=0.5,
        )
        self.play(Indicate(cells[5], color=accent), run_time=0.5)

        stack_vals.append(2)
        new_stack = redraw_stack()

        self.play(ReplacementTransform(stack_mobj, new_stack), run_time=0.7)
        stack_mobj = new_stack

        new_status = make_status("2 ≤ 4, άρα μπαίνει χωρίς pop")
        self.play(Transform(status, new_status), run_time=0.7)
        self.wait(1.8)

        summary_trace = Text(
            "Τελική στοίβα: 7, 6, 4, 2 — πάντα φθίνουσα από κάτω προς τα πάνω",
            font="DejaVu Sans",
            font_size=23,
            color=accent,
        )
        shrink_to_width(summary_trace, 10.5)
        summary_trace.move_to([0, -2.65, 0])

        self.play(FadeIn(summary_trace), run_time=0.7)
        self.wait(4.0)

        self.play(
            FadeOut(t3),
            FadeOut(arr_title),
            FadeOut(cells),
            FadeOut(labels),
            FadeOut(sf2),
            FadeOut(stack_mobj),
            FadeOut(current_marker),
            FadeOut(status),
            FadeOut(summary_trace),
            run_time=1.0,
        )

        # ==========================================================
        # 4) TEMPLATE
        # ==========================================================
        t4 = section_title(
            "template",
            "Ημιτελές skeleton — όχι πλήρης λύση συγκεκριμένου task",
        )
        self.play(FadeIn(t4, shift=UP), run_time=0.8)

        code_str = '''def solve(nums):
    stack = []
    ans = [None] * len(nums)   # ή άλλη δομή, ανάλογα με το task

    for i, x in enumerate(nums):
        while stack and stack[-1] <= x:
            # TODO:
            # - pop()
            # - ίσως ενημέρωση απάντησης / συσχέτιση index
            stack.pop()

        # TODO:
        # - διάβασε την κορυφή αν τη χρειάζεσαι
        # - αποθήκευσε τιμή ή index, ανάλογα με το ζητούμενο
        stack.append(x)

    return ans  # placeholder
'''

        code = Code(
            code_string=code_str,
            tab_width=4,
            language="Python",
            background="window",
            formatter_style="monokai",
        )

        shrink_to_width(code, 10.8)
        shrink_to_height(code, 5.05)
        code.move_to([0, -0.25, 0])

        note = Text(
            "Κράτα το μοτίβο, αλλά προσαρμόζεις τι αποθηκεύεις και πότε γράφεις απάντηση.",
            font="DejaVu Sans",
            font_size=23,
            color=accent,
        )
        place_bottom(note, y=STATUS_Y)

        self.play(FadeIn(code), run_time=1.0)
        self.play(FadeIn(note), run_time=0.8)
        self.wait(8.0)

        self.play(FadeOut(code), FadeOut(note), FadeOut(t4), run_time=1.0)

        # ==========================================================
        # 5) WARNINGS
        # ==========================================================
        t5 = section_title(
            "warnings",
            "Τι να προσέχω",
        )
        self.play(FadeIn(t5, shift=UP), run_time=0.8)

        warn_box = RoundedRectangle(
            corner_radius=0.15,
            width=11.0,
            height=4.55,
            color=bad,
        ).set_fill(bad, opacity=0.06)

        warn_box.shift(DOWN * 0.18)

        warns = VGroup(
            Text(
                "1. Τι βάζεις στη στοίβα; τιμές ή indices;",
                font="DejaVu Sans",
                font_size=26,
                color=WHITE,
            ),
            Text(
                "2. Αν έχεις ίσα στοιχεία, θες < ή <= στο while;",
                font="DejaVu Sans",
                font_size=26,
                color=WHITE,
            ),
            Text(
                "3. Η απάντηση γράφεται πριν ή μετά το push;",
                font="DejaVu Sans",
                font_size=26,
                color=WHITE,
            ),
            Text(
                "4. Ό,τι μπει βγαίνει το πολύ μία φορά → συνολικά O(n).",
                font="DejaVu Sans",
                font_size=26,
                color=WHITE,
            ),
        ).arrange(DOWN, aligned_edge=LEFT, buff=0.40)

        shrink_to_width(warns, 10.0)
        warns.move_to(warn_box.get_center())

        final_line = Text(
            "Στόχος: να αναγνωρίζεις το pattern, όχι να απομνημονεύεις έτοιμη λύση.",
            font="DejaVu Sans",
            font_size=24,
            color=accent,
        )
        place_bottom(final_line, y=STATUS_Y)

        self.play(Create(warn_box), run_time=0.8)

        for w in warns:
            self.play(Write(w), run_time=0.7)
            self.wait(1.2)

        self.play(FadeIn(final_line), run_time=0.8)
        self.wait(5.0)

        self.play(
            FadeOut(warn_box),
            FadeOut(warns),
            FadeOut(final_line),
            FadeOut(t5),
            run_time=1.0,
        )

        # -----------------------------
        # Outro
        # -----------------------------
        outro = VGroup(
            Text(
                "Φθίνουσα Μονοτονική Στοίβα",
                font="DejaVu Sans",
                font_size=42,
                weight=BOLD,
                color=title_color,
            ),
            Text(
                "Concept → Pattern → Trace → Template → Warnings",
                font="DejaVu Sans",
                font_size=24,
                color=text_color,
            ),
        ).arrange(DOWN, buff=0.25)

        shrink_to_width(outro, 11.2)

        self.play(FadeIn(outro, shift=UP), run_time=1.0)
        self.wait(3.0)
        self.play(FadeOut(outro), run_time=0.8)
```

```python
%%writefile build_talking_tutorial.py
from pathlib import Path
import subprocess
import shutil
import sys
import os

# =========================
# CONFIG
# =========================
ROOT = Path.cwd()

MANIM_FILE = ROOT / "monotonic_stack.py"
SCENE_NAME = "MonotonicStackQ1Tutorial"

# Βάλε εδώ την εικόνα avatar σου
AVATAR_IMAGE = ROOT / "avatar.png"

WORKDIR = ROOT / "talking_avatar_build"
WORKDIR.mkdir(exist_ok=True)

TUTORIAL_COPY = WORKDIR / "tutorial.mp4"
NARRATION_MP3 = WORKDIR / "narration.mp3"
NARRATION_WAV = WORKDIR / "narration.wav"

SADTALKER_DIR = ROOT / "SadTalker"
SADTALKER_RESULTS = WORKDIR / "sadtalker_results"
FINAL_VIDEO = WORKDIR / "final_tutorial_with_avatar.mp4"

# Μικρό μέγεθος avatar στην κάτω δεξιά γωνία
AVATAR_WIDTH = 320
OVERLAY_MARGIN_X = 24
OVERLAY_MARGIN_Y = 24

# =========================
# TUTORIAL NARRATION
# =========================
NARRATION_TEXT = """
Σε αυτό το tutorial θα δούμε τη φθίνουσα μονοτονική στοίβα.
Ο στόχος δεν είναι να πάρεις έτοιμη πλήρη λύση,
αλλά να καταλάβεις πώς σκέφτεσαι αυτό το pattern.

Πρώτα ξεκαθαρίζουμε τι ζητά η άσκηση.
Ποια είναι η είσοδος και ποια είναι η έξοδος.
Μετά ελέγχουμε αν το πρόβλημα ταιριάζει σε μονοτονική στοίβα.

Συνήθως το σκεφτόμαστε όταν ζητείται
επόμενο ή προηγούμενο μεγαλύτερο στοιχείο,
ή επόμενο ή προηγούμενο μικρότερο στοιχείο,
ή όταν θέλουμε να αφαιρούμε υποψήφιους
που δεν μπορούν πλέον να μας βοηθήσουν.

Η μονοτονική στοίβα δεν κρατά όλα τα στοιχεία του πίνακα.
Κρατά μόνο όσα παραμένουν χρήσιμα.
Στη φθίνουσα στοίβα,
από κάτω προς τα πάνω οι τιμές πρέπει να μένουν σε φθίνουσα σειρά.
Αν έρθει νέο στοιχείο που χαλάει αυτή τη σειρά,
τότε αφαιρούμε από την κορυφή όσα στοιχεία δεν είναι πια χρήσιμα.

Πριν γράψεις κώδικα πρέπει να πάρεις μερικές αποφάσεις.
Τι βάζεις στη στοίβα;
Τιμές, δείκτες ή και τα δύο;
Η στοίβα πρέπει να είναι αύξουσα ή φθίνουσα;
Πότε κάνεις pop;
Και πότε ενημερώνεις την απάντηση;

Αν χρειάζεσαι μόνο σύγκριση τιμών,
ίσως αρκεί να κρατάς τιμές.
Αν όμως χρειάζεσαι θέση, απόσταση
ή πρέπει να γράψεις αποτέλεσμα για συγκεκριμένο στοιχείο,
τότε συνήθως χρειάζεσαι δείκτες.

Η γενική ροή είναι η εξής.
Σαρώνεις τα στοιχεία με τη σωστή κατεύθυνση.
Όσο η κορυφή δεν ταιριάζει στον κανόνα,
κάνεις pop.
Ύστερα, αν το ζητά η άσκηση,
διαβάζεις την κορυφή ή ενημερώνεις την απάντηση.
Στο τέλος κάνεις push το τρέχον στοιχείο
ή τον δείκτη του.

Ας δούμε τώρα ένα παράδειγμα.
Η είσοδος είναι επτά, πέντε, έξι, τρία, τέσσερα, δύο.

Βλέπουμε πρώτα το επτά.
Η στοίβα είναι κενή, άρα το επτά μπαίνει.

Μετά βλέπουμε το πέντε.
Το πέντε είναι μικρότερο από το επτά,
οπότε δεν χαλάει τη φθίνουσα σειρά
και μπαίνει και αυτό στη στοίβα.

Τώρα βλέπουμε το έξι.
Η κορυφή είναι το πέντε.
Επειδή το έξι είναι μεγαλύτερο από το πέντε,
το πέντε δεν παραμένει χρήσιμο και αφαιρείται.
Μετά το pop, το έξι μπαίνει
και η στοίβα ξαναγίνεται φθίνουσα.

Μετά βλέπουμε το τρία.
Είναι μικρότερο ή ίσο από το έξι,
οπότε μπαίνει χωρίς pop.

Στη συνέχεια βλέπουμε το τέσσερα.
Η κορυφή είναι το τρία.
Επειδή το τέσσερα είναι μεγαλύτερο από το τρία,
το τρία αφαιρείται.
Έπειτα το τέσσερα μπαίνει πάνω από το έξι.

Τέλος βλέπουμε το δύο.
Το δύο είναι μικρότερο ή ίσο από το τέσσερα,
άρα μπαίνει χωρίς pop.

Στο τέλος η στοίβα είναι επτά, έξι, τέσσερα, δύο.
Παρατήρησε ότι δεν κρατά όλα τα στοιχεία της εισόδου.
Κρατά μόνο όσα δεν αφαιρέθηκαν από κάποιο μεταγενέστερο στοιχείο.

Ένα κρίσιμο σημείο είναι πότε ενημερώνεται η απάντηση.
Σε μερικές ασκήσεις αυτό γίνεται όταν γίνεται pop.
Σε άλλες, διαβάζεις την κορυφή πριν από το push.
Και σε άλλες χρειάζεσαι δείκτες
για να γράψεις το αποτέλεσμα στη σωστή θέση.

Προσοχή και στις ίσες τιμές.
Αν χρησιμοποιήσεις αυστηρή σύγκριση,
μπορεί να κρατήσεις ίσες τιμές στη στοίβα.
Αν χρησιμοποιήσεις μικρότερο ή ίσο,
τότε αφαιρείς και τις ίσες τιμές.
Η σωστή επιλογή εξαρτάται από το ζητούμενο.

Μην ξεχνάς και τα edge cases.
Ένα μόνο στοιχείο.
Ήδη αύξων πίνακας.
Ήδη φθίνων πίνακας.
Πολλές ίσες τιμές.
Και περιπτώσεις όπου ένα μεγάλο στοιχείο
προκαλεί πολλά pop.

Τέλος, γιατί η πολυπλοκότητα είναι γραμμική;
Παρότι βλέπεις while μέσα σε for,
κάθε στοιχείο μπαίνει στη στοίβα το πολύ μία φορά
και βγαίνει από τη στοίβα το πολύ μία φορά.
Άρα συνολικά τα push και pop είναι γραμμικά
και η πολυπλοκότητα είναι O του n.

Πριν υποβάλεις,
έλεγξε τι ακριβώς ζητά η άσκηση.
Επόμενο ή προηγούμενο;
Μεγαλύτερο ή μικρότερο;
Τι κρατάς στη στοίβα;
Τιμές ή δείκτες;
Τι κάνεις με ίσες τιμές;
Και πότε γράφεις το αποτέλεσμα;

Ο στόχος είναι να μπορείς να οδηγηθείς μόνος σου στη λύση,
χωρίς να απομνημονεύεις ένα έτοιμο template.
"""


# =========================
# HELPERS
# =========================
def run(cmd, cwd=None):
    if isinstance(cmd, list):
        printable = " ".join(str(x) for x in cmd)
    else:
        printable = cmd
    print(f"\n>>> {printable}\n")
    subprocess.run(cmd, cwd=cwd, check=True)


def require_file(path: Path, label: str):
    if not path.exists():
        raise FileNotFoundError(
            f"Δεν βρέθηκε το {label}: {path}\n"
            f"Βεβαιώσου ότι το αρχείο υπάρχει."
        )


def find_latest_file(pattern: str):
    files = list(ROOT.glob(pattern))
    if not files:
        return None
    files.sort(key=lambda p: p.stat().st_mtime)
    return files[-1]


# =========================
# STEP 1: RENDER MANIM
# =========================
def render_tutorial():
    require_file(MANIM_FILE, "monotonic_stack.py")

    run(["manim", "-pql", str(MANIM_FILE), SCENE_NAME])

    # Βρες το πιο πρόσφατο output video
    tutorial = find_latest_file("media/videos/**/MonotonicStackQ1Tutorial.mp4")
    if tutorial is None:
        raise FileNotFoundError(
            "Δεν βρέθηκε το tutorial mp4 μέσα στο media/videos."
        )

    shutil.copy2(tutorial, TUTORIAL_COPY)
    print(f"Tutorial video: {TUTORIAL_COPY}")
    return TUTORIAL_COPY


# =========================
# STEP 2: TTS AUDIO
# =========================
def generate_tts():
    from gtts import gTTS

    print("Δημιουργία TTS...")
    tts = gTTS(text=NARRATION_TEXT, lang="el")
    tts.save(str(NARRATION_MP3))

    # SadTalker δουλεύει καλύτερα με wav
    run([
        "ffmpeg", "-y",
        "-i", str(NARRATION_MP3),
        "-ar", "22050",
        "-ac", "1",
        str(NARRATION_WAV)
    ])

    print(f"TTS mp3: {NARRATION_MP3}")
    print(f"TTS wav: {NARRATION_WAV}")
    return NARRATION_WAV


# =========================
# STEP 3: SETUP SADTALKER
# =========================
def setup_sadtalker():
    if not SADTALKER_DIR.exists():
        run(["git", "clone", "https://github.com/OpenTalker/SadTalker.git"])

    marker = SADTALKER_DIR / ".deps_installed"
    if not marker.exists():
        run([sys.executable, "-m", "pip", "install", "-r", "requirements.txt"], cwd=SADTALKER_DIR)
        marker.write_text("ok", encoding="utf-8")

    checkpoints_dir = SADTALKER_DIR / "checkpoints"
    if not checkpoints_dir.exists():
        run(["bash", "scripts/download_models.sh"], cwd=SADTALKER_DIR)


# =========================
# STEP 4: CREATE TALKING AVATAR
# =========================
def create_talking_avatar():
    require_file(AVATAR_IMAGE, "avatar.png / avatar image")

    SADTALKER_RESULTS.mkdir(parents=True, exist_ok=True)

    run([
        sys.executable,
        "inference.py",
        "--driven_audio", str(NARRATION_WAV),
        "--source_image", str(AVATAR_IMAGE),
        "--result_dir", str(SADTALKER_RESULTS),
        "--still",
        "--preprocess", "full"
    ], cwd=SADTALKER_DIR)

    avatar_video = None
    mp4_files = list(SADTALKER_RESULTS.rglob("*.mp4"))
    if mp4_files:
        mp4_files.sort(key=lambda p: p.stat().st_mtime)
        avatar_video = mp4_files[-1]

    if avatar_video is None:
        raise FileNotFoundError("Δεν δημιουργήθηκε talking avatar video.")

    print(f"Talking avatar video: {avatar_video}")
    return avatar_video


# =========================
# STEP 5: OVERLAY AVATAR ON TUTORIAL
# =========================
def merge_videos(tutorial_video: Path, avatar_video: Path):
    run([
        "ffmpeg", "-y",
        "-i", str(tutorial_video),
        "-i", str(avatar_video),
        "-filter_complex",
        (
            f"[1:v]scale={AVATAR_WIDTH}:-1,"
            "tpad=stop_mode=clone:stop_duration=600[av];"
            "[0:v][av]overlay=W-w-24:H-h-24[v];"
            "[1:a]apad[a]"
        ),
        "-map", "[v]",
        "-map", "[a]",
        "-c:v", "libx264",
        "-c:a", "aac",
        "-shortest",
        str(FINAL_VIDEO)
    ])

    print(f"\n✅ Τελικό video: {FINAL_VIDEO}\n")
    return FINAL_VIDEO


# =========================
# MAIN
# =========================
def main():
    tutorial_video = render_tutorial()
    generate_tts()
    setup_sadtalker()
    avatar_video = create_talking_avatar()
    merge_videos(tutorial_video, avatar_video)


if __name__ == "__main__":
    main()
```
